## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#columns_to_drop = ['EIN', 'NAME']
c_t_drop = application_df.drop(['EIN'], axis=1)
c_t_drop.head()
```

```python
# Determine the number of unique values in each column.
c_t_drop.nunique()

#or
# for column in c_t_drop.columns:
#     ctd_uniq_vals = c_t_drop[column].nunique()

#     print(column, ctd_uniq_vals)
```

```python
# Look at NAME value counts for binning
name_cnts = c_t_drop['NAME'].value_counts()
name_cnts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# Chose my cutoff value
cutoff_value = 5

names_to_replace = name_cnts[name_cnts <= cutoff_value].index.tolist()

# Replace in dataframe
for app in names_to_replace:
    c_t_drop['NAME'] = c_t_drop['NAME'].replace(app,"Other")

# Check to make sure binning was successful
c_t_drop['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type_cnts = c_t_drop['APPLICATION_TYPE'].value_counts()
app_type_cnts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# Chose my cutoff value
cutoff_value = 500

application_types_to_replace = app_type_cnts[app_type_cnts < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    c_t_drop['APPLICATION_TYPE'] = c_t_drop['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
c_t_drop['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_vlu_cnts = c_t_drop.CLASSIFICATION.value_counts()
classification_vlu_cnts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
# Filter value counts greater than one
classification_vlu_cnts[classification_vlu_cnts > 1]
```

```python
classification_vlu_cnts[classification_vlu_cnts > 10]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#cutoff_value = 800

classifications_to_replace = classification_vlu_cnts[classification_vlu_cnts < 1000].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    c_t_drop['CLASSIFICATION'] = c_t_drop['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
c_t_drop['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# Convert categorical data to numeric using one-hot encoding
df_cat_to_num = pd.get_dummies(c_t_drop)
df_cat_to_num
```

```python
#predict whether applicants will be successful if funded by Alphabet Soup.
# Split our preprocessed data into our features and target arrays. 21.Class2.Act6.getting_real_solution.ipynb
y = df_cat_to_num["IS_SUCCESSFUL"].values
X = df_cat_to_num.drop(["IS_SUCCESSFUL", "STATUS"], axis='columns').values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  90
hidden_nodes_layer2 = 30
hidden_nodes_layer3 = 10

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="sigmoid"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```
